```python
import geopandas as gpd
```

```python
prov = gpd.read_file("provincias.json")

from shapely import geometry

points = [[-73.4154357571, -55.25], [-53.628348965, -55.25],[ -53.628348965, -21.8323104794],[-73.4154357571, -21.8323104794],]

poly = geometry.Polygon([[p[0], p[1]] for p in points])
df = gpd.GeoDataFrame(gpd.GeoSeries(poly), columns=['geometry'])
prov = gpd.overlay(prov, df, how='intersection')
```

```python
malvinas = [[-65, -53], [-55, -53],[ -55, -50],[-65, -50]]
malvinas = geometry.Polygon([[p[0], p[1]] for p in malvinas])

df = gpd.GeoDataFrame(gpd.GeoSeries(malvinas), columns=['geometry'])
malvinas = gpd.overlay(prov, df, how='intersection')
malvinas.loc[0, 'nombre'] = 'Islas Malvinas'
```

```python
prov = gpd.overlay(prov, malvinas, how = 'difference')
```

```python
import pandas as pd
prov = gpd.GeoDataFrame(pd.concat([prov, malvinas], axis=0))
prov.loc[prov.nombre.str.contains("Tierra"), 'nombre'] = "Tierra del Fuego"
```

```python
prov.to_file("provincias.geojson", driver='GeoJSON')
```
